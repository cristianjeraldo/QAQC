```python
# Librerías visualización
import matplotlib.pyplot as plt
import seaborn as sns

# Librerías manipulación de datos
import numpy as np
import pandas as pd
```

# Importación de datos

```python
# Importación de datos
df = pd.read_csv('Cu_sesgos.csv')
df
```

```python
analito = 'Cu'
```

```python
ax = df.plot(x='CRM', y=['ALS Lima','AAA Chile'], kind='bar', grid=None, ylabel='Sesgo', xlabel='', edgecolor='black')
plt.legend(['ALS Lima','AAA Chile'], loc='lower left')
plt.title('Comparación de sesgos por CRM para {}'.format(analito))
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.ylim(-10,10)
#plt.yticks(-10,2);
```
